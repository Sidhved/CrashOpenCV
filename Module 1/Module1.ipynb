# Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

Display Image Directly

```python
Image(filename="Zendaya.jpg")
```

Reading Images using OpenCV

```python
im = cv2.imread("Zendaya.jpg", 0)

print(im)   #numpy 2d column

plt.imshow(im)  #using matplotlib
```

```python
plt.imshow(im, cmap='gray')
```

Read and display color image

```python
im1 = cv2.imread("Zendaya.jpg", 1)

print("Image size: ", im1.shape)

print("Data type of image: ", im1.dtype)

plt.imshow(im1)
```

```python
im1_rev = im1[:, :, ::-1]
plt.imshow(im1_rev)
```

Splitting and Merging Color Channels

```python
im2 = cv2.imread("Zendaya.jpg", cv2.IMREAD_COLOR)
b, g, r = cv2.split(im2)

plt.figure(figsize=[20, 5])
plt.subplot(141); plt.imshow(r, cmap='gray'); plt.title("Red Channel");
plt.subplot(142); plt.imshow(g, cmap='gray'); plt.title("Green Channel");
plt.subplot(143); plt.imshow(b, cmap='gray'); plt.title("Blue Channel");

imMerged = cv2.merge((b, g, r))

plt.subplot(144); plt.imshow(imMerged[:, :, ::-1]); plt.title("Merged Output");
```

Converting to different Color Spaces

```python
im3 = cv2.cvtColor(im2, cv2.COLOR_BGR2RGB)
plt.imshow(im3)
```

Changing to HSV color space

```python
im4 = cv2.cvtColor(im2, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(im4)

plt.figure(figsize=[20, 5])
plt.subplot(141); plt.imshow(h, cmap='gray'); plt.title("H Channel");
plt.subplot(142); plt.imshow(s, cmap='gray'); plt.title("S Channel");
plt.subplot(143); plt.imshow(v, cmap='gray'); plt.title("V Channel");
plt.subplot(144); plt.imshow(im3, cmap='gray'); plt.title("Original Image");
```

Modifying Individual Channel

```python
h_new = h-20
im4_merge = cv2.merge((h_new, s, v))
im4_rgb = cv2.cvtColor(im4_merge, cv2.COLOR_HSV2RGB)

plt.figure(figsize=[20, 5])
plt.subplot(141); plt.imshow(h, cmap='gray'); plt.title("H Channel");
plt.subplot(142); plt.imshow(s, cmap='gray'); plt.title("S Channel");
plt.subplot(143); plt.imshow(v, cmap='gray'); plt.title("V Channel");
plt.subplot(144); plt.imshow(im4_rgb, cmap='gray'); plt.title("Modified");
```

Saving Images

```python
cv2.imwrite("Zendaya_saved.jpg", im4_rgb)

Image(filename="Zendaya_saved.jpg")
```
